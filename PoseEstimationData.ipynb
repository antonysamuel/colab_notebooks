```python
import pandas as pd
```

```python
df = pd.read_csv('./poseData/dataset/mpii_dataset.csv')
```

```python
carp_arr = []
mas_arr = []
conc_arr = []
plumb_arr = []
elect_arr = []
excav_arr = []
florr_arr = []
```

```python
img_name = df['NAME'].values
img_name[:10]
```

```python
actv = df['Activity'].values
actv[:10]
```

```python
len(img_name),len(actv)
```

```python
for a in range(len(actv)):
  wrd = actv[a].lower()
  aa = img_name[a]
  if 'carpentry' in wrd:
    carp_arr.append(aa)
  elif 'wiring' in wrd or 'electric' in wrd:
    elect_arr.append(aa)
  elif 'masonry' in wrd or 'concrete' in wrd:
    mas_arr.append(aa)
  elif 'concrete' in wrd:
    conc_arr.append(aa)
  elif 'mining' in wrd or 'digging' in wrd:
    excav_arr.append(aa)
  elif 'plumbing' in wrd:
    plumb_arr.append(aa)
  elif 'laying' in wrd:
    florr_arr.append(aa)
```

```python
len(carp_arr) + len(elect_arr)+len(mas_arr)+len(conc_arr) + len(excav_arr)+len(plumb_arr)+len(florr_arr) 
```
